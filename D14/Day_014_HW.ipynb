```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '../DATA/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
def col_2group(data,col_1,col_2):
    unique_house_type = data[col_1].unique()
    HOUSETYPE_MODE_gy = data.groupby(col_1)[col_2]
    nrows = len(unique_house_type)
    ncols = nrows/(nrows/2 )

    plt.figure(figsize=(20,30))
    for i in range(len(unique_house_type)):
        plt.subplot(nrows, ncols, i+1)
        plt.hist(data[col_2][data[col_1] == unique_house_type[i]])
        plt.title(str(unique_house_type[i]))
    supti_word = col_1+" group by "+col_2
    plt.suptitle(supti_word,size=32,y=0.92)
    plt.show()
col_2group(app_train,'HOUSETYPE_MODE','AMT_CREDIT')
```
